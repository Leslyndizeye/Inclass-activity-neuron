# Ganza Owen Yhaan & Ndizeye Lesly

Write a class Neuron that defines a single neuron performing binary classification:

class constructor: def __init__(self, nx):
nx is the number of input features to the neuron
If nx is not an integer, raise a TypeError with the exception: nx must be an integer
If nx is less than 1, raise a ValueError with the exception: nx must be a positive integer
All exceptions should be raised in the order listed above
Public instance attributes:
W: The weights vector for the neuron. Upon instantiation, it should be initialized using a random normal distribution.
b: The bias for the neuron. Upon instantiation, it should be initialized to 0.
A: The activated output of the neuron (prediction). Upon instantiation, it should be initialized to 0.

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv')
data.head(5)
```

```python
new_data = pd.get_dummies(data)
new_data.head(5)
```

```python
input = np.array(new_data.drop(['Legendary'], axis = 1), dtype = 'float')
print(input)
```

```python
print(input.shape)
```

```python
nx = input.shape[1]
```

```python
class Neuron:
    def __init__(self, nx):
        if not isinstance(nx, int):
            raise TypeError("nx must be an integer")
        if nx < 1:
            raise ValueError("nx must be a positive integer")

        self.W = np.random.randn(1, nx)
        self.b = 0.0
        self.A = 0.0

    def lin_reg(self, X):
      # Calculate z using linear equation mx = b
      return np.dot(self.W, X) + self.b

    def sigmoid(self, Z):
      # Squeeze the output to be btn 0-1
      return 1 / (1 + np.exp(-Z))

    def forward_propagate(self, X):
      Z = self.lin_reg(X)
      self.A = self.sigmoid(Z)
      return self.A
```

```python
# Transpose input for compatibility with neuron computation
X_test = input.T
# Reshape labels to match expected dimensions (1, number of examples)
Y_test = np.array(new_data['Legendary'].astype(int)).reshape(1, X_test.shape[1])

# Determine number of test examples
num_examples = X_test.shape[1]

print("--- Preparing Test Data ---")
print(f"X_test shape (features x examples): {X_test.shape}")
print(f"Y_test shape (1 x examples): {Y_test.shape}")
print(f"Input feature count (nx): {nx}")
print(f"Total number of test examples (m): {num_examples}")
print("-" * 40)

# Step 2: Initialize Neuron
try:
    print("--- Creating Neuron Instance ---")
    my_neuron = Neuron(nx=nx)  # Assumes nx is predefined
    print("Neuron created successfully.")
    print(f"Initial weights shape: {my_neuron.W.shape}")
    print(f"Initial bias value: {my_neuron.b}")
    print(f"Initial activation value: {my_neuron.A}")
    print("-" * 40)

    # Step 3: Generate predictions using forward propagation
    print("--- Running Forward Propagation ---")
    A_predictions = my_neuron.forward_propagate(X_test)
    print(f"Prediction output shape: {A_predictions.shape}")
    print("Preview of first 5 predicted probabilities:")
    print(A_predictions[0, :5])
    print("-" * 40)

    # Step 4: Convert probabilities into binary output using threshold
    print("--- Converting to Binary Predictions ---")
    binary_preds = (A_predictions >= 0.5).astype(int)
    print("First 5 predicted binary values:")
    print(binary_preds[0, :5])
    print("First 5 actual label values:")
    print(Y_test[0, :5])
    print("-" * 40)

    # Step 5: Evaluate basic prediction accuracy (prior to training)
    print("--- Evaluating Initial Accuracy ---")
    accuracy = np.mean(binary_preds == Y_test) * 100
    print(f"Untrained neuron accuracy: {accuracy:.2f}%")

except TypeError as te:
    print(f"TypeError encountered: {te}")
except ValueError as ve:
    print(f"ValueError encountered: {ve}")
except Exception as e:
    print(f"Unexpected error occurred: {e}")
```
